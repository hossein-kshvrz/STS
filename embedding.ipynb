```python
import collections
import datetime
import re
from hazm import *

FILE_NAME = 'data/fas_newscrawl_2017_100K/fas_newscrawl_2017_100K-sentences.txt'



def read_input(input_file):
    with open(input_file, encoding='utf-8') as f:
        content = f.readlines()
        return content


begin = datetime.datetime.now()
raw_content = read_input(FILE_NAME)
end = datetime.datetime.now()

```

```python
print(end-begin)
normalizer = Normalizer()


def my_normalizer(content):
    for i in range(len(content)):
        sentence = content[i]
        # remove number of sentence and \t
        sentence = sentence[sentence.index('\t')+1:]
        # hazm normalizer
        normalizer.normalize(sentence)
        # my tokenizer
        sentence = list(filter(lambda s: s != '', re.compile('[ /\'"،؛ء–«»():\-_.$,\[\]!؟\n\t]').split(sentence)))
        content[i] = sentence
#     content = [token for sentence in content for token in sentence]
    return content

document = my_normalizer(raw_content)
```

```python
print(document[0])
print(document[1])
print(document[2])
print(document[3])
```

```python
import gensim
# build vocabulary and train model
model = gensim.models.Word2Vec(
    document,
    size=300,
    window=5,
    sg=0,
    workers=10)
```

```python
model.train(document, total_examples=len(document), epochs=100)
```

```python
print(model.wv['تهران'])  # numpy vector of a word
result = model.wv.most_similar(positive=['تهران', 'عراق'], negative=['ایران'])
print(result)
result = model.wv.most_similar(positive=['دانشگاه'], topn=8)
print(result)
```

```python
selected_words = ['انتشار', 'جشنواره', 'پروژه', 'آژانس',
                  'فروردین', 'جام', 'رسانه', 'رابطه', 'مشکلات',
                  'رئیس', 'سياسی', 'روزنامه', 'دانشگاه']
for word in selected_words:
    print(word)
    print(model.wv.most_similar(positive=[word], topn=8))
    print(model.wv[word])
```
